```python
import pandas as pd
# Question answered: What is the full dataset we are analyzing for farmer support calls?
# What this tells us: Loading one combined file creates a single baseline for comparing
# support demand across states and understanding overall query coverage.
df = pd.read_csv("MP, UP and Raj.csv")
print(df.shape)
print(list(df.columns))
```

```python
# Question answered: What do sample records look like in this dataset?
# What this tells us: Seeing the first rows helps confirm whether the fields look
# reasonable and whether transcript content appears usable for support analysis.

df.head()
```

```python
# Question answered: What is the structural quality of the dataset (types and completeness)?
# What this tells us: df.info() shows row counts, data types, and non-null values so we can
# assess whether the data is complete enough for reliable farmer-support conclusions.

df.info()
```

```python
# Question answered: Where are missing values concentrated, especially in KccAns?
# What this tells us: High missingness in KccAns indicates potential response gaps, meaning
# many farmers may not have received a recorded answer in the support system.

df.isnull().sum()
```

```python
# Question answered: What are farmers asking about most often?
# What this tells us: The most frequent QueryType labels reveal the biggest support-demand
# themes, helping prioritize advisory content and staffing focus.

df["QueryType"].value_counts().head(10)
```

```python
# Question answered: Which queries went unanswered and what were they about?
# What this tells us: Unanswered count/percentage estimates service gaps, and QueryType
# distribution shows which issue categories are most affected by missing responses.

unanswered_df = df[df["KccAns"].isnull()].copy()

unanswered_count = len(unanswered_df)
total_count = len(df)
unanswered_pct = (unanswered_count / total_count) * 100

print(f"Unanswered queries: {unanswered_count} / {total_count} ({unanswered_pct:.2f}%)")
print(unanswered_df["QueryType"].value_counts())
```

```python
# Question answered: Do farmers in different states call about different problems?
# What this tells us: Ranking StateName x QueryType counts highlights state-specific issue
# patterns and where localized support priorities may differ across India.

df.groupby(["StateName", "QueryType"]).size().sort_values(ascending=False)
```
